```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns

data = [1, 2, 2, 3, 4, 4, 4, 5, 6]

plt.hist(data, bins=3, edgecolor='black')
```

```python
random_numbers = np.random.rand(100000)


plt.hist(random_numbers, bins=50,  alpha=0.5, color='blue', edgecolor='white')

plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Histogram of Uniformly Distributed Random Numbers')
plt.grid(True)

plt.show()
```

```python
random_numbers = np.random.rand(100000)

plt.hist(random_numbers, bins=50,  alpha=0.5, color='blue', edgecolor='black')

plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Histogram of Uniformly Distributed Random Numbers')
plt.grid(True)

plt.show()
```

```python
SNormalValues = np.random.randn(10000)#mean=0, standard deviation=1

plt.hist(SNormalValues, bins=30, edgecolor='black')
plt.title('Histogram of Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
NormalValues = np.random.normal(100, 10, 10000)

plt.hist(NormalValues, bins=30, edgecolor='black')
plt.title('Histogram of Normal Distribution (mean=100, std=10)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
